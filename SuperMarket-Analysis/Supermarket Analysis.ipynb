# Análise de histórico de vendas de um supermercado

```python
import pandas as pd
```

```python
df = pd.read_csv('supermarket_sales - Sheet1.csv')
df
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
df.drop(df.columns[13:15],axis=1,inplace=True)
df
```

```python
quantity = df['Product line'].value_counts()
quantity
```

```python
quantity = df.groupby(['Product line'])['Quantity'].sum()
quantity
```

```python
rate = df.groupby(['City'])['Rating'].mean()
rate
```

```python
total = df['Total'].sum()
total
```

```python
df['Date'] = pd.to_datetime(arg=df['Date'], format = "%m/%d/%Y")
df['Hour'] = pd.to_datetime(df['Time']).dt.hour
```

```python
def create_period(row):
    if row['Hour'] >= 0 and row['Hour'] <= 4:
        val = 'Dawn'
    elif row['Hour'] >= 5 and row['Hour'] <= 11:
        val = 'Morning'
    elif row['Hour'] >=12 and row['Hour'] <=17:
        val = 'Afternoon'
    elif row['Hour'] >= 18 and row['Hour'] <= 23:
        val = 'Night'
    return val
```

```python
df['time_period'] = df.apply(create_period,axis=1)
df
```

```python
period = df.groupby('time_period')['Total'].mean()
period
```

```python
df.groupby('time_period')['Product line'].value_counts()
```
